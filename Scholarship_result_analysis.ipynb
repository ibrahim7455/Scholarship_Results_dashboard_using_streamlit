```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r'C:\Users\Test\Desktop\projects\files\python\scholarship success and drop out\scholarships_data.csv')
```

```python
df.head()
```

```python
df['IQ_test_score'] = np.ceil(df['IQ_test_score'])
df['English_test_score'] = np.ceil(df['English_test_score'])
df['technical_test_score'] = np.ceil(df['technical_test_score'])
df['Soft_skills_Score'] = np.ceil(df['Soft_skills_Score'])
```

```python
df.head()
```

```python
df.info()
```

```python
df['IQ_test_score'] = df['IQ_test_score'].astype(int)
df['English_test_score'] = df['English_test_score'].astype(int)
df['technical_test_score'] = df['technical_test_score'].astype(int)
df['Soft_skills_Score'] = df['Soft_skills_Score'].astype(int)
```

```python
df.info()
```

```python
new_names = {
    'field': 'major',
    'drop out reason': 'drop_out_reason',
    'sub major':'sub_major'
}

df = df.rename(columns=new_names)
```

```python
df['gender'] = df['gender'].str.lower()
```

```python
df.isnull().sum()
```

```python
df['gender'].fillna('Male' , inplace=True )
df['ethnicity'].fillna(df['ethnicity'].mode()[0], inplace=True)
df['education_level'].fillna(df['education_level'].mode()[0], inplace=True)
```

```python
df['drop_out_reason'] = df['drop_out_reason'].fillna('No reason')
```

```python
df.isnull().sum()
```

```python
df.to_csv("processed_student_data.csv", index=False)
```

## EDA process

```python
plt.figure(figsize=(10, 6))

result_order = ['Accepted', 'pending', 'drop out']

ax = sns.countplot(data=df, x='Result', order=result_order,
                  saturation=0.9,
                  edgecolor='black', linewidth=0.7)

plt.title('Distribution of Final Results for Students', 
          fontsize=14, pad=20, fontweight='bold')
plt.xlabel('Final Result', fontsize=12, labelpad=10)
plt.ylabel('Number of Students', fontsize=12, labelpad=10)

for p in ax.patches:
    height = p.get_height()
    ax.annotate(f'{height:.0f}',  
               (p.get_x() + p.get_width() / 2., height),
               ha='center', va='bottom',
               fontsize=11,
               bbox=dict(boxstyle='round,pad=0.3', 
                        fc='white', ec='gray', alpha=0.8),
               xytext=(0, 5),  
               textcoords='offset points')

plt.axhline(y=0, color='gray', linestyle='-', linewidth=0.5)
plt.grid(axis='y', linestyle=':', alpha=0.4)

plt.tight_layout()
plt.show()
```

```python
cross_tab = pd.crosstab(df['financial_aid'], df['Result'], normalize='index') * 100

ax = cross_tab.plot(kind='bar', stacked=True, figsize=(10,6),
                   
                   edgecolor='black')

plt.title('Percentage Distribution of Results by Financial Aid', fontsize=14)
plt.xlabel('Financial Aid Status', fontsize=12)
plt.ylabel('Percentage (%)', fontsize=12)
plt.xticks(rotation=0)
plt.legend(title='Result', bbox_to_anchor=(1.05, 1))
plt.grid(axis='y', alpha=0.3)

for rect in ax.patches:
    height = rect.get_height()
    if height > 0:
        ax.text(rect.get_x() + rect.get_width()/2., 
                rect.get_y() + height/2.,
                f'{height:.1f}%', 
                ha='center', va='center',
                color='black', fontsize=10)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
result_order = ['Accepted', 'pending', 'drop out']

ax = sns.boxplot(data=df, x='Result', y='gpa', order=result_order,
                width=0.5, linewidth=1.5, fliersize=5)

plt.title('Distribution of GPA by Final Result', 
          fontsize=14, pad=20, fontweight='bold')
plt.xlabel('Final Result', fontsize=12, labelpad=10)
plt.ylabel('GPA', fontsize=12, labelpad=10)
plt.ylim(0, 4)  

sns.stripplot(data=df, x='Result', y='gpa', order=result_order,
              color='black', alpha=0.3, size=4, jitter=True)

plt.grid(axis='y', linestyle='--', alpha=0.4)

medians = df.groupby('Result')['gpa'].median().loc[result_order]
for i, median in enumerate(medians):
    ax.text(i, median, f'Median: {median:.2f}', 
            ha='center', va='bottom', fontsize=10,
            bbox=dict(boxstyle='round,pad=0.3', fc='white', ec='none', alpha=0.7))

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 7))

ax = sns.countplot(data=df, x='english_level', hue='Result',
                  palette='muted',  
                  order=sorted(df['english_level'].unique()),  
                  saturation=0.9,   
                  edgecolor='black', linewidth=0.5)  

plt.title('Distribution of Results by English Proficiency Level', fontsize=16, pad=20, fontweight='bold')
plt.xlabel('English Proficiency Level', fontsize=13, labelpad=10)
plt.ylabel('Number of Students', fontsize=13, labelpad=10)

plt.xticks(rotation=45, ha='right', fontsize=11)

for p in ax.patches:
    height = p.get_height()
    if height > 0:  
        ax.annotate(f'{int(height)}',
                   (p.get_x() + p.get_width() / 2., height),
                   ha='center', va='bottom',
                   fontsize=10,
                   xytext=(0, 5),  
                   textcoords='offset points',
                   bbox=dict(boxstyle='round,pad=0.2', fc='white', ec='none', alpha=0.7))

plt.legend(title='Final Result', title_fontsize=12, fontsize=11,
          bbox_to_anchor=(1.05, 1), loc='upper left',
          frameon=True, shadow=True, facecolor='white')

plt.grid(axis='y', linestyle='--', alpha=0.4)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
ax = sns.countplot(data=df, x='current_employment_status', hue='Result', 
                   palette='viridis', edgecolor='black', linewidth=0.5)

plt.title('Distribution of Results by Current Employment Status', fontsize=14, pad=20)
plt.xlabel('Employment Status', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(rotation=45, ha='right')

for p in ax.patches:
    height = p.get_height()
    if height > 0:  
        ax.annotate(f'{int(height)}', 
                   (p.get_x() + p.get_width() / 2., height),
                   ha='center', va='bottom',
                   fontsize=10,
                   xytext=(0, 3),
                   textcoords='offset points')

plt.legend(title='Result', bbox_to_anchor=(1.05, 1), loc='upper left', 
           frameon=True, shadow=True)

plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```
